```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dane=pd.read_csv('D:\\STUDIA\\ROK_II\\Magisterka\\Modele\\poprawne wyniki.txt',sep='\t')
```

```python
dane
```

## 1. M

```python
dane_m=dane[dane['class']=='M'].sort_values(by=['F1-score'], ascending=False)
```

```python
dane_m
```

```python
dane_m_f1=dane_m[dane_m['class']=='M']['F1-score']
dane_m_p=dane_m[dane_m['class']=='M']['Precission']
dane_m_r=dane_m[dane_m['class']=='M']['Recall']
dane_m_modele=dane_m['model']
```

```python
dane_m_f1
```

```python
# Numbers of pairs of bars you want
N = 6

# Data on X-axis

# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, dane_m_f1 , width, label='F1-score')
plt.bar(ind + width, dane_m_p, width, label='Precyzja')
plt.bar(ind + width+width, dane_m_r, width, label='Czułość')

plt.xlabel('Model')
plt.ylabel('Wartość')
plt.title('Klasa M - metryki wydajności')
plt.ylim(0, 1)
# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, dane_m_modele)

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```
